```python
import pandas as pd
```

```python
house = pd.read_csv("../data/lagos_houses.csv")
```

# Data Cleaning

```python
house.head()
```

```python
# house.astype({'Price': 'int64'}, errors='ignore')
house['Price'] = house['Price'].str.replace(',', '')
house['Price'] = pd.to_numeric(house["Price"], errors='coerce')
# house.dtypes
```

```python
house.describe()
```

```python
house.dropna(inplace=True)
house.head(20)
```

```python
house.describe()
```

```python
house = house[house["Beds"] != 0]
house.head(20)
```

```python
house.describe()
```

```python
house = house[(house["Baths"] != 0) & (house["Toilets"] != 0)]
house.head(20)
```

```python
house.describe()
```

# Feature Representation

```python
dummies = pd.get_dummies(house["Area"], prefix="Area")
dummies.head()
house.drop(columns=["Area"], inplace=True)
new_house = pd.concat([dummies, house], axis=1)
new_house.head(20)
```

```python
new_house.dtypes
new_house.head(20)
```

```python
new_house.to_csv("../data/clean_houses.csv", index=False)
```
